```python
import matplotlib.pyplot as plt

years = range(2000, 2019)
b=[]
for anne in range (2000,2019):
    b.append(chomagetaux(anne))

plt.plot(years, b)
plt.title(' ')
plt.xlabel('Année')
plt.ylabel('Taux de chomage (%)')
plt.show()
    
    
```

```python
population=[63174483, 64100297, 65022300, 65938265, 66845635, 67743052, 68626337, 69496513, 70363511, 71241080, 72137246, 73058638, 75627384, 76667864, 77695904, 78741053, 79814871, 80810525, 82003882, 83154997]


def calculate_growth_rate(previous_population, next_population):
    return ((next_population / previous_population) - 1) * 100


growth_rates = [calculate_growth_rate(population[i-1], population[i]) for i in range(1, len(population))]

print(growth_rates)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 5))
plt.style.use('ggplot')


plt.plot(range(2001, 2020), growth_rates, color='blue')


plt.xlabel('Anné')
plt.ylabel('Augmentation de taux de population(%)')
plt.title("Taux d'augmentation annuel de la population entre 2000 et 2019")


plt.show()
```
